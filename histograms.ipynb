```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
data = h5py.File('data.h5', 'r')
```

```python
number_of_labels = [len(v) for k, v in data['labels'].items()]
plt.xlabel("number of labels")
plt.ylabel("amount")
plt.hist(number_of_labels)
```

```python
number_by_class = dict()
for _img_id, labels in data['labels'].items():
    for label in labels:
        class_type = label[-1]
        if not class_type in number_by_class:
            number_by_class[class_type] = 1
        else:
            number_by_class[class_type] += 1

```

```python
plt.plot(number_by_class.values())
```

```python
l = []
for _img_id, labels in data['labels'].items():
    for label in labels:
        class_type = label[-1]
        l.append(class_type)

plt.hist(l)
```
